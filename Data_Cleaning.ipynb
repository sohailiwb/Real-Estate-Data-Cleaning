```python
## filling missing carpet_area with help of built up area


# Remove commas from 'super_built_up_area' and 'carpet_area'
df['super_built_up_area'] = df['super_built_up_area'].astype(str).str.replace(',', '')
df['carpet_area'] = df['carpet_area'].astype(str).str.replace(',', '')

#  Convert both columns to numeric, forcing errors to NaN
df['super_built_up_area'] = pd.to_numeric(df['super_built_up_area'], errors='coerce')
df['carpet_area'] = pd.to_numeric(df['carpet_area'], errors='coerce')

# Calculate efficiency factor (Carpet Area / Super Built-up Area)
df['efficiency_factor'] = df['carpet_area'] / df['super_built_up_area']

#  Compute estimated efficiency factor (ignoring NaN values)
estimated_factor = df['efficiency_factor'].mean()

#  Fill missing Carpet Area using estimated efficiency factor
df['carpet_area'] = df['carpet_area'].fillna(df['super_built_up_area'] * estimated_factor)

# Convert 'carpet_area' to integer type
df['carpet_area'] = df['carpet_area'].astype(int)

# Drop the temporary 'efficiency_factor' column (Optional)
df.drop(columns=['efficiency_factor'], inplace=True)


```

```python
df.isnull().sum()
```

```python
## converion of price in numeric value
def convert_price_to_inr(price):
    if isinstance(price, str):  # Ensure it's a string
        price = price.replace('₹', '').strip().lower()  # Remove currency symbol and spaces

        if 'cr' in price:
            return float(price.replace('cr', '').strip()) * 1_00_00_000  # Convert Cr to INR
        elif 'lac' in price:
            return float(price.replace('lac', '').strip()) * 1_00_000  # Convert Lac to INR
    
    return None  # Return None for invalid values

# Apply conversion
df['price_in_inr'] = df['Price'].apply(convert_price_to_inr)

# Drop rows where conversion failed (invalid values)
df = df.dropna(subset=['price_in_inr'])

# Reset index after dropping rows
df = df.reset_index(drop=True)
```

```python
##approx price per sqft of missing data 
df["price_per_sqft"] = (df["price_in_inr"] / df["super_built_up_area"]).astype(str) + " /sqft"
```

```python

# Define IQR range for super_built_up_area
Q1 = df["super_built_up_area"].quantile(0.25)
Q3 = df["super_built_up_area"].quantile(0.75)
IQR = Q3 - Q1

upper_bound = Q3 + 1.5 * IQR

# Filter the dataset
df = df[(df["super_built_up_area"] >=700) & (df["super_built_up_area"] <= upper_bound)]

#Define IQR range for price_per_sqft
Q1 = df["price_per_sqft"].quantile(0.25)
Q3 = df["price_per_sqft"].quantile(0.75)
IQR = Q3 - Q1

upper_bound = Q3 + 1.5 * IQR

# Filter the dataset
df = df[(df["price_per_sqft"] <= upper_bound)]
```
